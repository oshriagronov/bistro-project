```python
import paho.mqtt.client as mqtt
import json

# Callback function for when a message is received
def on_message(client, userdata, msg):
    try:
        data = json.loads(msg.payload.decode())

        # Extract values with lowercase field names
        temperature = data.get("temperature", "N/A")
        humidity = data.get("humidity", "N/A")
        soil= data.get("soil", "N/A")
        tempThreshold = data.get("tempThreshold", "N/A")
        humThreshold = data.get("humThreshold", "N/A")
        print(f"Temperature: {temperature}°C, Temperature Threshold: {tempThreshold}°C, Humidity: {humidity}%, Humidity Threshold: {humThreshold}°C")

    except json.JSONDecodeError:
        print("Received invalid JSON data")
        # MQTT setup
broker = "io.adafruit.com"
username = "braude2"   #insert Adafruit user name
aio_key = "" #insert key from Adafruit
topic = f"{username}/feeds/json"
client = mqtt.Client()
client.username_pw_set(username,aio_key)
client.on_message = on_message

client.connect(broker, 1883, 60)
client.subscribe(topic)

print(f"Subscribed to MQTT topic: {topic}")
client.loop_forever()

```

```python
import requests
import matplotlib.pyplot as plt
import pandas as pd
import json

USERNAME = "braude2"   #insert Adafruit user name
AIO_KEY = "" #insert key from Adafruit
FEED = "json"

url = f"https://io.adafruit.com/api/v2/{USERNAME}/feeds/{FEED}/data"
headers = {"X-AIO-Key": AIO_KEY}

# Lists to hold the parsed data
timestamps = []
temperatures = []
humidities = []

response = requests.get(url, headers=headers)
data = response.json()
# Loop through the last 100 items from the previously fetched 'data'
for item in data[:100]:
    try:
        # Parse the 'value' string which contains the JSON metrics
        val_dict = json.loads(item['value'])

        timestamps.append(pd.to_datetime(item['created_at']))
        temperatures.append(val_dict.get('temperature'))
        humidities.append(val_dict.get('humidity'))
    except (json.JSONDecodeError, KeyError, TypeError):
        continue

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(timestamps, temperatures, label='Temperature (°C)', color='red', marker='o', markersize=2)
plt.plot(timestamps, humidities, label='Humidity (%)', color='blue', marker='o', markersize=2)

plt.title('Temperature and Humidity (Last 100 Readings)')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.legend()
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
